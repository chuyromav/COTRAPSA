```python
import pandas as pd
import pygal
from datetime import datetime
import time
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('/Users/Laura/Desktop/COTRAPSA/Analisis facturas/Facturas-convertido.xlsx')
```

```python
df.shape
```

```python
df.head()
```

```python
df['fh_entr'] = pd.to_datetime(df.Fecha_de_entr.astype(str)).dt.date
df['Periodo_pago']=df.Fecha_de_p-df.Fecha_de_elabor

df['Fecha_de_elabor']=pd.to_datetime(df['Fecha_de_elabor'].astype(str)).dt.date
df['Periodo_entr']=(df.Fecha_de_elabor-df.fh_entr)
```

```python
df=df[df.Estatus=='Emitida']
df.reset_index(inplace=True,drop=True)
```

```python
Num_facturas=df[['Nombre','Importe']].groupby('Nombre').count()
Num_facturas.sort_values('Importe',ascending=False,inplace=True)
```

```python
df['Año']=df['Fecha_de_elabor'].map(lambda x:x.strftime('%Y'))
df['Año-mes']=df['Fecha_de_p'].map(lambda x:x.strftime('%Y%m'))
df['mes']=df['Fecha_de_p'].map(lambda x:x.strftime('%m'))
```

```python
df2019=df[df['Año']=='2019']
df2018=df[df['Año']=='2018']
df2017=df[df['Año']=='2017']
```

```python
Num_facturas_2019=df2019[['Nombre','Importe']].groupby('Nombre').count()
Num_facturas_2019.sort_values('Importe',ascending=False,inplace=True)

Num_facturas_2018=df2018[['Nombre','Importe']].groupby('Nombre').count()
Num_facturas_2018.sort_values('Importe',ascending=False,inplace=True)

Num_facturas_2017=df2017[['Nombre','Importe']].groupby('Nombre').count()
Num_facturas_2017.sort_values('Importe',ascending=False,inplace=True)
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Facturas x Cliente'
line_chart.x_labels = Num_facturas_2019.index
line_chart.add('2019',Num_facturas_2019.Importe)
line_chart.add('2018',Num_facturas_2018.Importe)
line_chart.add('2017',Num_facturas_2017.Importe)
line_chart.add('2016-2019',Num_facturas.Importe)
line_chart.render_to_file('Facturas_x_cliente.svg')
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Facturas x Cliente'
line_chart.x_labels = Num_facturas_2019.index
line_chart.add('2019',Num_facturas_2019.Importe)
line_chart.render_to_file('Facturas_x_cliente_2019.svg')
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Facturas x Cliente'
line_chart.x_labels = Num_facturas_2018.index
line_chart.add('2019',Num_facturas_2019.Importe)
line_chart.add('2018',Num_facturas_2018.Importe)
line_chart.render_to_file('Facturas_x_cliente_2018.svg')
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Facturas x Cliente'
line_chart.x_labels = Num_facturas_2017.index
line_chart.add('2017',Num_facturas_2017.Importe)
line_chart.render_to_file('Facturas_x_cliente_2017.svg')
```

```python
pie_chart = pygal.Pie()
pie_chart.title = 'Facturas x Cliente'
pie_chart.add('2019',Num_facturas_2019.Importe)
pie_chart.add('2018',Num_facturas_2018.Importe)
pie_chart.add('2017',Num_facturas_2017.Importe)
pie_chart.render_to_file('Facturas_x_cliente_pie.svg')
```

```python
mensual_2019=df2019[['mes','Importe']].groupby('mes').count()
mensual_2018=df2018[['mes','Importe']].groupby('mes').count()
mensual_2017=df2017[['mes','Importe']].groupby('mes').count()
```

```python
line_chart = pygal.HorizontalBar()
line_chart.title = 'Facturas x Cliente al mes 2019'
line_chart.x_labels = mensual_2018.index
line_chart.add('2019',mensual_2019.Importe)
line_chart.add('2018',mensual_2018.Importe)
line_chart.add('2017',mensual_2017.Importe)
line_chart.render_to_file('Facturas_x_cliente_mensual_2019.svg')
```

```python
mensual2_2019=df2019[['mes','Importe','Fecha_de_elabor']].groupby('mes').sum()
mensual2_2018=df2018[['mes','Importe','Fecha_de_elabor']].groupby('mes').sum()
mensual2_2017=df2017[['mes','Importe','Fecha_de_elabor']].groupby('mes').sum()

mensual2_2019['fecha_importe'] = list(zip(mensual2_2019.index.astype(int), mensual2_2019.Importe.astype(int)))
mensual2_2018['fecha_importe'] = list(zip(mensual2_2018.index.astype(int), mensual2_2018.Importe.astype(int)))
mensual2_2017['fecha_importe'] = list(zip(mensual2_2017.index.astype(int), mensual2_2017.Importe.astype(int)))
```
